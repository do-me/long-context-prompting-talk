```python
import pandas as pd 
df = pd.read_excel("https://github.com/do-me/SDG/raw/refs/heads/main/SDG_Target_Indicator_UNSD_2023.xlsx")
df
```

```python
with open("sdg_indicators.txt", "w") as file:
    for index, row in df.iterrows():
        for col in ["SDGTitle", "IndicatorsNumber", "Indicators"]:
            file.write(f"{str(row[col]).strip()}\n")
        file.write("---\n")
```
